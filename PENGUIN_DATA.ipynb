```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('penguins_size.csv')


data = data.dropna()


plt.figure(figsize=(10, 6))
plt.scatter(data['culmen_length_mm'], data['body_mass_g'], c=data['sex'].apply(lambda x: 0 if x == 'MALE' else 1), cmap='viridis', label='Sex')
plt.colorbar(label='Sex (0=MALE, 1=FEMALE)')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Penguin Culmen Length vs Body Mass')
plt.legend()
plt.show()
```
